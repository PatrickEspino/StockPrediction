# Predicting Stock Prices
Models/Simulation will be trained on the first 80% of the data and tested on the last 20%

1. Monte Carlo
2. Deep Learning Model (LSTM)

```python
from scipy.stats import norm

def split_data(df, split = 0.8):
    train = df.head(int(len(df) * split)).copy()
    test = df.iloc[train.shape[0]:df.shape[0]]
    return train, test

def monte_carlo(df, iterations):
    train, test = split_data(df)
    train['log_returns'] = train['daily_return'].apply(lambda x : np.log(1 + x))
    u = train['log_returns'].mean()
    var = train['log_returns'].var()
    drift = u - (0.5 * var)
    stdv = train['log_returns'].std()
    days = test.shape[0]
    np.random.seed(1)
    Z = norm.ppf(np.random.rand(days, iterations))
    daily_returns = np.exp(np.array(drift) + np.array(stdv * Z))
    starting_price = train.iloc[-1]
    price_list = np.zeros_like(daily_returns)
    price_list[0] = starting_price['Adj Close']
    for i in range(1, days):
        price_list[i] = price_list[i - 1] * daily_returns[i]
    price_list_df = pd.DataFrame(price_list)
    differences = np.array([])
    for i in range(0, iterations):
        difference_arrays = np.subtract(price_list_df[i].values, test['Adj Close'])
        difference_values = np.sum(np.abs(difference_arrays))
        differences = np.append(differences, difference_values)
    best_fit = np.argmin(differences)
    
    return train, test, daily_returns, price_list, best_fit

def predict_future(daily_returns, best_fit_df, days):
    predict = best_fit_df.squeeze()
    for i in range(0, days):
        predict.loc[len(predict.index)] = (predict.loc[len(predict.index) - 1] * daily_returns[i])[best_fit_df.columns[0]]
    predict_df = pd.DataFrame(predict)
    return predict_df
```

```python
train, test, daily_returns, price, best = monte_carlo(GOOG, 1000)
```

```python
plt.figure(figsize = (12,5))
plt.plot(price)
plt.grid(color = 'k', linestyle = '-.', linewidth = 0.5)
plt.title('Monte Carlo Simulation of Google Stock Price')
plt.xlabel('Days')
plt.ylabel('Price in USD')
plt.show()
```

```python
#Compare best simulation to actual price
price_list_df = pd.DataFrame(price)
best_fit_df = price_list_df[best].to_frame()
test_df = test.reset_index()
comparison_df = pd.concat([best_fit_df, test_df['Adj Close']], axis = 1)

plt.figure(figsize = (12,5))
plt.plot(comparison_df)
plt.grid(color = 'k', linestyle = '-.', linewidth = 0.5)
plt.title('Best Fit vs. Actual Price of Google Stock')
plt.legend(labels = ['Predicted', 'Actual'])
plt.xlabel('Days')
plt.ylabel('Price in USD')
plt.show()
```

```python
#Predict Future Stock Price of Google
predict_df = predict_future(daily_returns, best_fit_df, 10)
future_df = pd.concat([predict_df, best_fit_df, test_df['Adj Close']], axis = 1)
future_df
```

```python
import matplotlib.ticker as ticker
tick_spacing = 5

plt.figure(figsize = (12,5))
plt.plot(future_df)
plt.grid(color = 'k', linestyle = '-.', linewidth = 0.5)
plt.legend(labels = ['Future', 'Predicted', 'Actual'])
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.title('Stock Prediction for Google')
plt.xlabel('Days')
plt.ylabel('Price in USD')
plt.show()
```

# Implementing LSTM

Using Adj Close

```python
data = GOOG['Adj Close'].reset_index(drop = True)
data.head()
```

```python
train, test = split_data(data)
data = pd.concat([train, test])
data
```

```python
y = data.values.reshape(-1, 1)
y
```

```python
#Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
scaler = scaler.fit(y)
y = scaler.transform(y)
y
```

```python
n_lookback = 25
n_forecast = 10

X = []
Y = []

for i in range(n_lookback, len(y) - n_forecast + 1):
    X.append(y[i - n_lookback: i])
    Y.append(y[i: i + n_forecast])

X = np.array(X)
Y = np.array(Y)
```

```python
#Create LSTM
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

model = Sequential()
model.add(LSTM(units = 25, return_sequences = True, input_shape = (n_lookback, 1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 25, return_sequences = True))
model.add(Dropout(0.2))

model.add(LSTM(units  = 25, return_sequences = True))
model.add(Dropout(0.2))

model.add(LSTM(units = 50))
model.add(Dense(n_forecast))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(X,Y, epochs = 100, batch_size = 16, verbose = 0)
```

```python
X_ = y[-n_lookback:]
X_ = X_.reshape(1, n_lookback, 1)

Y_ = model.predict(X_).reshape(-1,1)
Y_ = scaler.inverse_transform(Y_)
```

```python
Y_.shape
```

```python
predicted_df = pd.DataFrame(Y_)
predicted_df.rename(columns={0 : 'Predicted'}, inplace = True)
data_df = pd.DataFrame(test).reset_index(drop = True)

prediction_df = pd.concat([data_df, predicted_df], ignore_index = True)
prediction_df
```

```python
#Show prediction of the next 10 days
plt.figure(figsize = (12,5))
plt.plot(prediction_df)
plt.grid(color = 'k', linestyle = '-.', linewidth = 0.5)
plt.legend(labels = ['Actual', 'Prediction'])
plt.xlabel('Days')
plt.ylabel('Price in USD')
plt.title('Actual Stock and Forecasted Price')
plt.show()
```
